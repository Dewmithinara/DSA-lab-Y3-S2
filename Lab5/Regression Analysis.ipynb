# Multiple linear regression

```python
data = pd.read_csv("Startups.csv")
```

```python
data['State'] = data['State'].factorize()[0]  #factorize
```

```python
y = data.iloc[:,4] # ll rows, 4th columns
```

```python
x1 =data.iloc[:,0:1] #independent variable
```

```python
x2 =data.iloc[:,2:3] #Marketing spend
```

```python
x3 =data.iloc[:,1:2] #administration
```

```python
x4 =data.iloc[:,3:4] #state
```

# plot graphs

```python
plt.scatter(x1,y) # independent x , dependent y
plt.title('R&D spend vs profit ') # plot title
plt.xlabel('R&D spend' ) # x axis
plt.ylabel('profit') # y axis
plt.show() 
```

```python
plt.scatter(x2,y) # independent x , dependent y
plt.title('administration vs Year of profit') # plot title
plt.xlabel('administration') # x axis
plt.ylabel('profit') # y axis
plt.show() 
```

```python
plt.scatter(x3,y) # independent x , dependent y
plt.title('marketing spend  vs profit') # plot title
plt.xlabel('marketing spend') # x axis
plt.ylabel('profit') # y axis
plt.show() 
```

```python
plt.scatter(x4,y) # independent x , dependent y
plt.title('state vs profit') # plot title
plt.xlabel('state') # x axis
plt.ylabel('profit') # y axis
plt.show() 
```

```python
# R&D Spend vs Profit
X = sm.add_constant(x1)  # add intercept
model1 = sm.OLS(y, X).fit()
print("R&D Spend vs Profit:\n", model1.summary())
```

```python
# Marketing Spend vs Profit
X = sm.add_constant(x2)
model2 = sm.OLS(y, X).fit()
print("Marketing Spend vs Profit:\n", model2.summary())
```

```python
# Admin Spend vs Profit
X = sm.add_constant(x3)
model3 = sm.OLS(y, X).fit()
print("Admin Spend vs Profit:\n", model3.summary())
```

```python
# State vs Profit
X = sm.add_constant(x4)
model4 = sm.OLS(y, X).fit()
print("State vs Profit:\n", model4.summary())
```
